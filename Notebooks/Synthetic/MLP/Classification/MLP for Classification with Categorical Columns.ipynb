```python
X = df.drop('loan_approval', axis=1)
y = df['loan_approval']


```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_test_original = X_test.copy() 
```

```python
loan_mapping = {'Small': 2, 'Medium': 1, 'Large': 0}
income_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
education_mapping = {'High School': 0, 'Bachelor': 1, 'Master': 2}

# Function to apply mappings to both train and test
def apply_category_mappings(df):
    df = df.copy()  # avoid changing original

    # Manual mappings
    df["Loan_Amount"] = df["Loan_Amount"].map(loan_mapping)
    df["Income_Level"] = df["Income_Level"].map(income_mapping)
    df["Education_Level"] = df["Education_Level"].map(education_mapping)

    # Encode remaining categorical columns
    for col in features:
        if col not in ["Loan_Amount", "Income_Level", "Education_Level"]:
            df[col] = df[col].astype('category').cat.codes

    return df
```

```python
X_train_encoded = apply_category_mappings(X_train)
X_test_encoded = apply_category_mappings(X_test)
```

```python
def compute_feature_recovery_metrics(gt, ranking):
    """
    Compute TP, FP, FN, Precision, Recall, FDR, and Top-1 Match for feature recovery.

    Parameters
    ----------
    gt : set
        Ground-truth causal features.
    flagged : list or set
        Features your method flagged as important (e.g., topâ€‘k from ranking).
    ranking : list
        Ordered list of features by descending effect size.

    Returns
    -------
    pd.DataFrame
        Single-row DataFrame with the metrics.
    """
    tp = set(gt) & set(ranking)
    fp = set(ranking) - set(gt)
    fn = set(gt) - set(ranking)

    precision = len(tp) / (len(tp) + len(fp)) if (len(tp) + len(fp)) else 0.0
    recall    = len(tp) / len(gt)        if len(gt)              else 0.0
    fdr       = len(fp) / (len(tp) + len(fp)) if (len(tp) + len(fp)) else 0.0
    top1      = 1 if ranking and ranking[0] in gt else 0

    metrics = {
        "TP":           ", ".join(sorted(tp)),
        "FP":           ", ".join(sorted(fp)),
        "FN":           ", ".join(sorted(fn)),
        "Precision":     precision,
        "Recall":        recall,
        "FDR":           fdr,
        "Top-1 Match":   top1
    }
    return pd.DataFrame([{
        "Precision": precision,
        "Recall": recall,
        "FDR": fdr,
        "Top-1 Match": top1
    }])



if __name__ == "__main__":
    # Example:
    GT = {"Loan_Amount", "Credit_History", "Income_Level"}
    ranking = [
        "Loan_Amount",
        "Credit_History",
        "annual_income",
        "Income_Level",
        "Education_Level",
        "Zip_Code"
    ]
    # flagged = ranking[:3]  # e.g. topâ€‘3 features

    df_metrics = compute_feature_recovery_metrics(GT, ranking)
    df_metrics.insert(0, "Method", "SHAP-XAI")
    df_metrics = df_metrics[["Method", "Precision", "Recall", "FDR", "Top-1 Match"]]
    print(df_metrics.to_string(index=False))
```

```python
explainer = shap.Explainer(lambda x: clf.predict_proba(x)[:, 1], X_train_encoded)
```

```python
batch_size = 2000
all_shap_values = []
```

```python
for start in range(0, X_test_encoded.shape[0], batch_size):
    end = start + batch_size
    X_batch = X_test_encoded.iloc[start:end]

    # Check memory before each batch
    mem = psutil.virtual_memory()
    print(f"Batch {start}-{end:,}: Available RAM = {mem.available/1e9:.2f}â€¯GB")

    # Compute SHAP for this batch (each call returns shape [batch_size, n_features])
    shap_batch = explainer(X_batch)
    all_shap_values.append(shap_batch.values)
```

```python
all_shap_values = np.vstack(all_shap_values)
```

```python
n_plot = 2000
idx_sample = np.random.choice(all_shap_values.shape[0], size=n_plot, replace=False)
X_plot = X_test_encoded.iloc[idx_sample]
shap_vals_plot = all_shap_values[idx_sample]   # now already the classâ€‘1 SHAP values

# 4. Summary plot
shap.summary_plot(shap_vals_plot, X_plot)
```

```python
import os
os.makedirs("shap", exist_ok=True)

plt.figure(figsize=(6,4))
shap.summary_plot(
    shap_vals_plot,
    X_plot,
    feature_names=X.columns,
    plot_type="bar",
    show=False      # turn off autoâ€‘show so we can fineâ€‘tune with plt
)
plt.title("Mean |SHAP value| per feature")
plt.tight_layout()
# plt.savefig("synthetic categorical and classification.jpg", format="jpg", dpi=300)
output_path = os.path.join("shap", "synthetic categorical and classification.jpg.jpg")
plt.savefig(output_path, format="jpg", dpi=300)
plt.show()
```
